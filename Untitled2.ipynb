```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# =======================
# Input Data
# =======================
data = """--


"""

df = pd.read_csv(StringIO(data))

# =======================
# 1. Throughput vs Message Size (Workers = 4)
# =======================
plt.figure()
subset = df[df["Workers"] == 4]
for impl in subset["Impl"].unique():
    d = subset[subset["Impl"] == impl]
    plt.plot(d["Size"], d["ThroughputGbps"], marker='o', label=impl)

plt.xlabel("Message Size (bytes)")
plt.ylabel("Throughput (Gbps)")
plt.title("Throughput vs Message Size")
plt.legend()
plt.show()

# =======================
# 2. Latency vs Worker Count (Size = 4096)
# =======================
plt.figure()
subset = df[df["Size"] == 4096]
for impl in subset["Impl"].unique():
    d = subset[subset["Impl"] == impl]
    plt.plot(d["Workers"], d["LatencyUs"], marker='o', label=impl)

plt.xlabel("Workers")
plt.ylabel("Latency (Âµs)")
plt.title("Latency vs Worker Count")
plt.legend()
plt.show()

# =======================
# 3. Cache Misses vs Message Size (LLC Misses, Workers = 4)
# =======================
plt.figure()
subset = df[df["Workers"] == 4]
for impl in subset["Impl"].unique():
    d = subset[subset["Impl"] == impl]
    plt.plot(d["Size"], d["LLCMiss"], marker='o', label=impl)

plt.xlabel("Message Size (bytes)")
plt.ylabel("LLC Misses")
plt.title("LLC Cache Misses vs Message Size")
plt.legend()
plt.show()

# =======================
# 4. CPU Cycles per Byte Transferred (Workers = 4)
# =======================
df["CyclesPerByte"] = df["Cycles"] / df["Size"]

plt.figure()
subset = df[df["Workers"] == 4]
for impl in subset["Impl"].unique():
    d = subset[subset["Impl"] == impl]
    plt.plot(d["Size"], d["CyclesPerByte"], marker='o', label=impl)

plt.xlabel("Message Size (bytes)")
plt.ylabel("Cycles per Byte")
plt.title("CPU Cycles per Byte Transferred")
plt.legend()
plt.show()
```
